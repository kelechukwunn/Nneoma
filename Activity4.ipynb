# Business Problem

Jade Fresneda is the Director of Analytics at a large car insurance company on the West Coast. Jade has hired your analytics consulting company to assist with data storytelling related to personal auto loan customers.

A basic overview of the **PersonalCarIns.csv** data is below:

| Variable |Description |
| ----------- | ----------- |
|Customer| unique customer identification number|
|State| customer's state of residence|
|Coverage| customer's level of insurance coverage (Basic, Extended, Premium)|
|Education| level of education of insured (High School or Below, College, Bachelor, Master, Doctor)|
|Employment| employment status of customer at time of inception (Employed, Unemployed, Medical Leave, Disabled, Retired)|
|Gender| gender of customer (F: Female; M: Male)|
|Income| yearly income of customer|
|Location Code| code identifying the location type of customer's residence (Suburban, Rural, Urban)|
|Marital Status| customer's marital status (Single, Married, Divorced)|
|Monthly Premium Auto| amount paid per month (USD)|
|Months Since Last Claim| the number of months since the customer's last insurance claim|
|Months Since Policy Inception| the number of months since the custoemr's insurance policy was established|
|Number of Open Complaints| the number of open insurance complaints|
|Number of Policies| the number of insurance policies|
|Sales Channel| sales channel of policy inception (Agent, Branch, Call Center, Web)|
|Total Claim Amount| amount of claims in USD|
|Vehicle Class| vehicle class (Sports Car, Two-Door Car, Four-Door Car, SUV, Luxury Car, Luxury SUV)|
|Vehicle Size| size of vehicle insured (Small, Medsize, Large)|
